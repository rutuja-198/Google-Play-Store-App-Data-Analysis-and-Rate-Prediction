# Cleaning Data

Dataset can contain missing data, numerical string value, duplication etc.

## 1. Category

```python
#fetch unique value of category
data['Category'].unique()
```

```python
#fetch the category which is  1.9
data[data['Category'] == '1.9']
```

```python
#shift the category to NaN using shift function
data.loc[10472] = data.loc[10472].shift()
data['App'].loc[10472] = data['Category'].loc[10472]
data['Category'].loc[10472] = np.nan
data.loc[10472]
```

## 2. Rating

```python
# check data
data['Rating'].unique()
```

```python
#datatype is object so convert into numeric
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')
data['Rating'].dtype
```

## 3. Review

```python
#check data
data['Reviews'].unique()
```

```python
#replace the review in million and check data type
data['Reviews'] = data.Reviews.replace("0.0",0)
data['Reviews'] = data.Reviews.replace("3.0M",3000000.0)
```

```python
data['Reviews'] = data['Reviews'].astype(float)
data['Reviews'].dtype
```

## 4. Size 

```python
#check data
data['Size'].unique()
```

```python
data['Size'] = data.Size.replace("Varies with device",np.nan) #changed 'Varies with device' value with Nan
data['Size'] = data.Size.str.replace("M","000") # All size values became the kilobyte type.
data['Size'] = data.Size.str.replace("k","") 
data['Size'] = data.Size.replace("1,000+",1000)
data['Size'] =data['Size'].astype(float) #converted float value
data['Size'].dtype
```

## 5. Installs

```python
#check data
data['Installs'].unique()
```

```python
#replace the strings and convert data type
data['Installs'] = data.Installs.str.replace(",","")
data['Installs'] = data.Installs.str.replace("+","")
data['Installs'] = data.Installs.replace("Free",np.nan)
data['Installs'] = data['Installs'].astype(float)
data['Installs'].dtype
```

## 6. Price 

```python
#check data
data['Price'].unique()
```

```python
data['Price'] = data.Price.replace("Everyone",np.nan)
data['Price'] = data.Price.str.replace("$","").astype(float)
data['Price'].dtype
```

## 7. Last Updated 

```python
#check data
data['Last Updated'].unique()
```

```python
#convert data type
data['Last Updated'] = pd.to_datetime(data['Last Updated'])
data['Last Updated']
```

## 8. Apps

```python
#check data
data['App'].unique()
```

```python
#check any duplicate values
duplicate_val = data['App'].duplicated().any()
duplicate_val
```

```python
data['App'].value_counts()
```

#### As we can see from the above mentioned data, some of the apps are having multiple rows.

## Exploratory Data Analysis

Let’s plot a visualization graph to view what portion of the apps in the play store are paid and free.

### 1. Percentage of Free vs Paid Apps in store 

```python
plt.figure(figsize=(10,10))
labels = data['Type'].value_counts(sort = True).index
sizes = data['Type'].value_counts(sort = True)
colors = ["blue","lightgreen"]
explode = (0.2,0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=0)
plt.title('Percent of Free Vs Paid Apps in store',size = 20)
plt.show()
```

#### It can be conclude that the 92.6% is free application in play store and 7.4% is paid application in play store.

### 2. Distribution of Rating 

```python
#distribution of the ratings
plt.figure(figsize=(15,9))
plt.xlabel("Rating")
plt.ylabel("Frequency")
graph = sns.kdeplot(data.Rating, color="Blue", shade = True)
plt.title('Distribution of Rating',size = 20);
```

### 3. Count of applications in each category differentiated by their type

```python
count_of_apps = data.groupby(['Category','Type'])[['App']].count().reset_index().rename(columns={'App':'Count','index':'App'})
df_app_count = count_of_apps.pivot('Category', 'Type', 'Count').fillna(0).reset_index()
df_app_count.set_index('Category').plot(kind='bar', stacked=True, figsize=(18,9))
plt.xlabel("Category", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Count of applications in each category differentiated by their type")
plt.show()
```

#### It looks like certain app categories have more free apps available for download than others. In dataset, the majority of apps in Family, Food & Drink and Tools, as well as Social categories were free to install and at the same time Family, Sports, Tools and Medical categories had the biggest number of paid apps available for download.

### 4. Total of installed apps according to its type

```python
categoty_type = data.groupby(['Category','Type'])[['Installs']].sum().reset_index()
categoty_type['log_Installed'] = np.log2(categoty_type['Installs'])
plt.figure(figsize = (18,9))
plt.xticks(rotation = 65,fontsize = 9)
plt.xlabel("Category")
plt.ylabel("Installs")
plt.title("Number of installs type wise according to Category")
sns.barplot('Category', 'log_Installed', hue='Type', data=categoty_type);
plt.show()
```

#### It can be conclude that the number of free applications installed by the user are high when compared with the paid ones.

## Analysis on Google Play Store Review

```python
user_review_data = pd.read_csv("C:/Users/Admin/Desktop/Google Play Store Analysis/Google Play Store Apps Dataset/googleplaystore_user_reviews.csv")
```

```python
#merge dataframes on apps
merged_data = data.merge(user_review_data, on="App")
```

### Distribution of type of reviews in the dataset

```python
category_sentiment = merged_data.groupby(['Category','Sentiment']).size().reset_index(name='Sentiment Count')
```

```python
category_sentiment['log_sentiment_count'] = np.log2(category_sentiment['Sentiment Count'])
```

```python
plt.figure(figsize=(18,9))
plt.xticks(rotation=90,fontsize=11)
plt.xlabel("Category", fontsize=15)
plt.ylabel("Installs", fontsize=15)
plt.title("Number of installs type wise according to Genres",fontsize=15)
sns.barplot('Category', 'log_sentiment_count', hue='Sentiment', data=category_sentiment);
```

### Percentage of Review Sentiments

```python
import matplotlib
counts = list(merged_data['Sentiment'].value_counts())
labels = 'Positive Reviews', 'Negetive Reviews','Neutral Reviews'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (10, 10)
plt.pie(counts, labels=labels, explode=[0, 0.05, 0.005], shadow=True, autopct="%.2f%%")
plt.title('A Pie Chart Representing Percentage of Review Sentiments', fontsize=20)
plt.axis('off')
plt.legend()
plt.show()
```
